# **Images for prompting**

```python
import PIL.Image
img = PIL.Image.open('/content/schedule.png')
img
```

```python
prompt = """This image contains a timetable of technical session along with Day/Date, timing and topics of the session.
Given the timetable images, describe that anyone wants to attend the technical session must have to follow this timetable then and then only they will be able to join the session on the regular basis.
see in the image, making sure to note all of the session topics and timings as well as day/date too. Return output in json format:
{description: description, features: [feature1, feature2, feature3, etc]}"""
```

```python
model = genai.GenerativeModel('gemini-pro-vision')
response = model.generate_content([prompt, img])
print(response.text)
```
